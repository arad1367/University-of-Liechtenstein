### 3. Basic Plotting with Matplotlib



#### 3.1 Line Plots

* Line plots are used to visualize trends over time or continuous data.

```python
# Example: Line Plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='Sine Wave')
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

#### 3.2 Scatter Plots

* Scatter plots are useful for visualizing relationships between two variables.

```python
# Example: Scatter Plot
np.random.seed(19680801)  # Seed the random number generator
data = {
    'a': np.arange(50),
    'c': np.random.randint(0, 50, 50),
    'd': np.random.randn(50)
}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.scatter('a', 'b', c='c', s='d', data=data)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
plt.show()
```

#### 3.3 Bar Plots

* Bar plots are ideal for comparing categorical data.

```python
# Example: Bar Plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

plt.bar(categories, values, color='skyblue')
plt.title("Bar Plot Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

### 4. Customizing Plots



#### 4.1 Titles, Labels, and Legends

* Adding titles, axis labels, and legends improves the interpretability of your plots.

```python
# Example: Adding Titles and Labels
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, x, label='linear')  # Linear plot
ax.plot(x, x**2, label='quadratic')  # Quadratic plot
ax.plot(x, x**3, label='cubic')  # Cubic plot
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title("Customized Plot")
ax.legend()
plt.show()
```

#### 4.2 Colors, Markers, and Line Styles

* Customize the appearance of your plots using colors, markers, and line styles.

```python
# Example: Customizing Appearance
fig, ax = plt.subplots(figsize=(5, 2.7))
x = np.arange(len(data['a']))
ax.plot(x, np.cumsum(data['a']), color='blue', linewidth=3, linestyle='--')
l, = ax.plot(x, np.cumsum(data['b']), color='orange', linewidth=2)
l.set_linestyle(':')
plt.show()
```

#### 4.3 Grids and Annotations

* Grids and annotations enhance readability and provide additional context.

```python
# Example: Adding Grids and Annotations
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.grid(True)  # Enable grid
plt.annotate('Max Point', xy=(np.pi / 2, 1), xytext=(2, 0.5),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

### 5. Subplots and Multiple Axes

* Subplots allow you to display multiple plots in a single figure.

```python
# Example: Subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 6), layout='constrained')

# Scatter plot
axs[0, 0].scatter(data['a'], data['b'], c=data['c'], s=data['d'])
axs[0, 0].set_title("Scatter Plot")

# Line plot
x = np.linspace(0, 2, 100)
axs[0, 1].plot(x, x**2, label='Quadratic')
axs[0, 1].set_title("Line Plot")
axs[0, 1].legend()

# Bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]
axs[1, 0].bar(categories, values, color='skyblue')
axs[1, 0].set_title("Bar Plot")

# Histogram
axs[1, 1].hist(np.random.randn(1000), bins=30, color='purple', edgecolor='black')
axs[1, 1].set_title("Histogram")

plt.show()
```

### 7. Advanced Plotting Examples

* Here are some advanced examples showcasing the flexibility of Matplotlib.

```python
# Example: Contour and Heatmap Plots
X, Y = np.meshgrid(np.linspace(-3, 3, 128), np.linspace(-3, 3, 128))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

fig, axs = plt.subplots(2, 2, layout='constrained')
pc = axs[0, 0].pcolormesh(X, Y, Z, vmin=-1, vmax=1, cmap='RdBu_r')
fig.colorbar(pc, ax=axs[0, 0])
axs[0, 0].set_title('pcolormesh()')

co = axs[0, 1].contourf(X, Y, Z, levels=np.linspace(-1.25, 1.25, 11))
fig.colorbar(co, ax=axs[0, 1])
axs[0, 1].set_title('contourf()')

pc = axs[1, 0].imshow(Z**2 * 100, cmap='plasma')
fig.colorbar(pc, ax=axs[1, 0], extend='both')
axs[1, 0].set_title('imshow()')

pc = axs[1, 1].scatter(data['a'], data['b'], c=data['c'], cmap='RdBu_r')
fig.colorbar(pc, ax=axs[1, 1], extend='both')
axs[1, 1].set_title('scatter()')

plt.show()
```
