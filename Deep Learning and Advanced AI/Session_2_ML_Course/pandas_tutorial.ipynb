### 1. Introduction to Pandas

> Pandas is a powerful Python library designed for data manipulation and analysis. It provides easy-to-use data structures and tools for working with structured data, making it an essential tool for data scientists and analysts.

Key Features:
- Handles missing data efficiently.
- Supports reading and writing data from various file formats (e.g., CSV, Excel, SQL).
- Provides tools for reshaping, merging, slicing, and indexing data.
- Integrates well with other libraries like NumPy and Matplotlib.

### 2. Installation and Import

```python
# Install Pandas (if not already installed) -- In colab, you don't need to install pandas
# !pip install pandas

# Import Pandas
import pandas as pd
print(f"Pandas version: {pd.__version__}")
```

### 3. Core Data Structures in Pandas

* A `Series` is a one-dimensional labeled array capable of holding any data type.
* A `DataFrame` is a two-dimensional labeled data structure with columns of potentially different types.

```python
# Series

# Creating a Series
data = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd']) # index is not compulsory
print(data)
```

```python
# Creating a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}
df = pd.DataFrame(data)
print(df)
```

### 4. Loading Real Datasets

```python
# Load Employee Data
employee_data_url = "https://raw.githubusercontent.com/arad1367/WAC_November-2023/refs/heads/main/Employee_Data_missing.csv"
employee_df = pd.read_csv(employee_data_url)
print("Employee Data:")
print(employee_df.head())

# Load Drug Data
drug_data_url = "https://raw.githubusercontent.com/arad1367/WAC_November-2023/refs/heads/main/drug200_missing.csv"
drug_df = pd.read_csv(drug_data_url)
print("\nDrug Data:")
print(drug_df.head())
```

### 5. Data Exploration

* Exploring data is the first step in any data analysis project. Pandas provides several methods to quickly understand your dataset.

```python
# Display basic information about the dataset
print(employee_df.info())
print("\nDrug Data Info:")
print(drug_df.info())

# Summary statistics
print("\nEmployee Data Summary:")
print(employee_df.describe())
print("\nDrug Data Summary:")
print(drug_df.describe())

# First 5 rows
print("\nFirst 5 rows of Employee Data:")
print(employee_df.head())
```

### 6. Data Cleaning

* Real-world datasets often contain missing or inconsistent data. Pandas provides tools to handle these issues.

```python
# Check for missing values
print("Missing Values in Employee Data:")
print(employee_df.isnull().sum())

# Drop rows with missing values
cleaned_employee_df = employee_df.dropna()
print("\nCleaned Employee Data:")
print(cleaned_employee_df)

# Fill missing values
filled_drug_df = drug_df.fillna({'Cholesterol': 'Unknown', 'Na_to_K': 0})
print("\nFilled Drug Data:")
print(filled_drug_df)
```

### 7. Data Manipulation

* Pandas allows you to manipulate data efficiently using indexing, filtering, and transformations.

```python
# Filter employees with Salary > 70000
high_salary_employees = employee_df[employee_df['Salary'] > 70000]
print(high_salary_employees)

# Add a new column: Bonus (10% of Salary)
employee_df['Bonus'] = employee_df['Salary'] * 0.1
print("\nEmployee Data with Bonus Column:")
print(employee_df)

# Remove the Bonus column
employee_df = employee_df.drop('Bonus', axis=1)
print("\nEmployee Data after Removing Bonus Column:")
print(employee_df)
```

### 8. GroupBy Operations

* GroupBy operations allow you to aggregate data based on specific criteria.

```python
# Group employees by Department and calculate average salary
avg_salary_by_dept = employee_df.groupby('Department')['Salary'].mean()
print(avg_salary_by_dept)
```

### 9. Merging and Joining DataFrames

* Combine multiple DataFrames using merge or join operations.

```python
# Create two DataFrames
df1 = pd.DataFrame({'Key': ['A', 'B', 'C'], 'Value': [1, 2, 3]})
df2 = pd.DataFrame({'Key': ['B', 'C', 'D'], 'Value': [4, 5, 6]})

# Merge DataFrames on 'Key'
merged_df = pd.merge(df1, df2, on='Key', how='inner')
print(merged_df)
```

### 10. Visualization with Pandas

* Pandas integrates with Matplotlib for quick visualizations.

```python
# Plot a bar chart of average salary by department
avg_salary_by_dept.plot(kind='bar', title='Average Salary by Department')
```
