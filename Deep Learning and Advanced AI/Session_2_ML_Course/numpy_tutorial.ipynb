### 1. Introduction to NumPy

NumPy (short for **Numerical Python**) is one of the most fundamental libraries in Python for scientific computing. It provides support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays efficiently.  

NumPy serves as the foundation for many other libraries in the Python ecosystem, such as **Pandas, SciPy, TensorFlow, and PyTorch**.

#### Why Use NumPy?

- **🚀 Efficiency:** NumPy operations are implemented in optimized C code, making them significantly faster than native Python loops.  
- **📏 Flexibility:** It supports multidimensional arrays and a wide range of mathematical operations.  
- **🔗 Integration:** It integrates seamlessly with other libraries used in data science and machine learning.  
- **💾 Memory Efficiency:** NumPy arrays consume less memory compared to Python lists.

### 3. Core Concepts in NumPy

#### 3.1 Arrays

The core data structure in NumPy is the `ndarray` (n-dimensional array). Unlike Python lists, NumPy arrays are homogeneous (all elements must be of the same type) and are stored in contiguous memory blocks, which makes them highly efficient.

#### 3.2 Array Attributes

Every NumPy array has several important attributes

```python
# Example of Array Attributes
arr = np.array([[1, 2, 3], [4, 5, 6]])

print("Array:\n", arr)
print("Shape:", arr.shape)           # Shape of the array
print("Number of dimensions:", arr.ndim)  # Number of dimensions
print("Data type:", arr.dtype)       # Data type of elements
print("Size:", arr.size)             # Total number of elements
print("Item size (in bytes):", arr.itemsize)  # Size of each element in bytes
```

### 4. Creating NumPy Arrays

#### 4.1 From Python Lists

```python
# Create a NumPy array from a Python list
arr = np.array([1, 2, 3, 4, 5])
print(arr)
```

#### 4.2 Using Built-in Functions

```python
# Zeros Array
zeros_arr = np.zeros((3, 4))  # Creates a 3x4 array filled with zeros
print(zeros_arr)

# Ones Array
ones_arr = np.ones((2, 3))  # Creates a 2x3 array filled with ones
print(ones_arr)

# Identity Matrix
identity_matrix = np.eye(4)  # Creates a 4x4 identity matrix
print(identity_matrix)

# Random Array
random_arr = np.random.random((2, 3))  # Creates a 2x3 array with random values between 0 and 1
print(random_arr)

# Arange
range_arr = np.arange(0, 10, 2)  # Creates an array with values from 0 to 8 with step size 2
print(range_arr)

# Linspace
linspace_arr = np.linspace(0, 1, 5)  # Creates 5 evenly spaced values between 0 and 1
print(linspace_arr)
```

### 5. Indexing and Slicing

```python
# 1D Array
arr = np.array([1, 2, 3, 4, 5])

print(arr[0])      # First element
print(arr[-1])     # Last element
print(arr[1:4])    # Elements from index 1 to 3

# 2D Array
arr = np.array([[1, 2, 3], [4, 5, 6]])

print(arr[0, 1])   # Element in the first row, second column
print(arr[:, 1])   # All rows, second column
print(arr[1, :])   # Second row, all columns
```

### 6. Basic Operations



#### 6.1 Arithmetic Operations

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

print("Addition:", arr1 + arr2)
print("Subtraction:", arr1 - arr2)
print("Multiplication:", arr1 * arr2)
print("Division:", arr1 / arr2)
```

#### 6.2 Broadcasting

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
scalar = 2

print("Broadcasted Multiplication:\n", arr * scalar)
```

### 7. Universal Functions (ufuncs)

```python
arr = np.array([1, 2, 3, 4])

print("Square Root:", np.sqrt(arr))
print("Exponential:", np.exp(arr))
print("Logarithm:", np.log(arr))
print("Sine:", np.sin(arr))
```

### 8. Aggregation Functions

```python
arr = np.array([1, 2, 3, 4, 5])

print("Sum:", np.sum(arr))
print("Mean:", np.mean(arr))
print("Minimum:", np.min(arr))
print("Maximum:", np.max(arr))
print("Standard Deviation:", np.std(arr))
```

### 9. Reshaping and Manipulating Arrays


#### 9.1 Reshaping

```python
arr = np.arange(12)
reshaped = arr.reshape((3, 4))  # Reshape into a 3x4 array
print(reshaped)
```

#### 9.2 Concatenation and Splitting

```python
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6]])
concatenated = np.concatenate((arr1, arr2), axis=0)
print(concatenated)

arr = np.arange(10)
split_arrays = np.split(arr, 2)  # Split into 2 equal parts
print(split_arrays)
```

### 10. Linear Algebra

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
result = np.matmul(A, B)
print(result)

arr = np.array([[1, 2], [3, 4]])
print("Transpose:\n", arr.T)

arr = np.array([[1, 2], [3, 4]])
inverse = np.linalg.inv(arr)
print("Inverse:\n", inverse)
```

### 11. Working with Images in NumPy

* Images are essentially multidimensional arrays of pixel values. In this section, we will explore how to load an image, convert it into a NumPy array, and manipulate it using NumPy operations.

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import requests
```

```python
# Load the image from the URL
image_url = "https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcSMrTWz33b86nfIrgaW9jE_t-7VCcqJtjL-pg&s"
image = Image.open(requests.get(image_url, stream=True).raw)
print(image)

# Convert the image to a NumPy array
image_array = np.array(image)

print("Image Shape:", image_array.shape)  # Shape: (height, width, channels)
print("Data Type:", image_array.dtype)   # Data type of pixel values
```

```python
# Display the image
plt.imshow(image_array)
plt.axis('off')  # Turn off axis labels
plt.title("Original Image")
plt.show()
```

```python
# Check if the image has 3 dimensions (height, width, color channels)
if image_array.ndim == 3:  # RGB or RGBA image
    # Convert to grayscale by averaging along the color channels (axis=2)
    gray_image = np.mean(image_array, axis=2).astype(np.uint8)
else:  # Already grayscale (2D array)
    gray_image = image_array.astype(np.uint8)

# Display the grayscale image
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.title("Grayscale Image")
plt.show()
```

```python
# Crop the image (e.g., top-left corner)
cropped_image = image_array[:100, :100]  # First 100 rows and columns

# Display the cropped image
plt.imshow(cropped_image)
plt.axis('off')
plt.title("Cropped Image")
plt.show()
```

```python
# Resize the image to half its size
resized_image = image.resize((image.width // 2, image.height // 2))
resized_array = np.array(resized_image)

# Display the resized image
plt.imshow(resized_array)
plt.axis('off')
plt.title("Resized Image")
plt.show()
```

```python
# Check if the image has 3 dimensions (height, width, color channels)
if image_array.ndim == 3:  # RGB or RGBA image
    # Extract the red channel
    red_channel = image_array.copy()
    red_channel[:, :, 1] = 0  # Set green channel to 0
    red_channel[:, :, 2] = 0  # Set blue channel to 0
else:  # Already grayscale (2D array)
    print("The image is already grayscale and has no color channels.")
    red_channel = image_array  # Use the grayscale image as-is

# Display the filtered image
plt.imshow(red_channel)
plt.axis('off')
plt.title("Red Channel Only")
plt.show()
```
